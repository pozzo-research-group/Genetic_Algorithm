```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Best Estimate Concentrations

```python
for itr in range(1,7):
    PW_BE_SPHERE = np.load('../Data_peakwv_SPHERE/' + str(itr) + '/best_estimate.npy')
    Euclidean_BE_SPHERE = np.load('../Data_euclidean_SPHERE/' + str(itr) + '/best_estimate.npy')
    AP_BE_SPHERE = np.load('../Data_AP_SPHERE/' + str(itr) + '/best_estimate.npy')
    SRVF_BE_SPHERE = np.load('../Data_SRVF_SPHERE/' + str(itr) + '/best_estimate.npy')
    PW_BE_ROD = np.load('../Data_peakwv_ROD/' + str(itr) + '/best_estimate.npy')
    Euclidean_BE_ROD = np.load('../Data_euclidean_ROD/' + str(itr) + '/best_estimate.npy')
    AP_BE_ROD = np.load('../Data_AP_ROD/' + str(itr) + '/best_estimate.npy')
    SRVF_BE_ROD = np.load('../Data_SRVF_ROD/' + str(itr) + '/best_estimate.npy')
    if itr == 1:
        PW_ROD_ALL = PW_BE_ROD
        Euclidean_ROD_ALL = Euclidean_BE_ROD
        SRVF_ROD_ALL = SRVF_BE_ROD
        AP_ROD_ALL = AP_BE_ROD
        PW_SPHERE_ALL = PW_BE_SPHERE
        Euclidean_SPHERE_ALL = Euclidean_BE_SPHERE
        SRVF_SPHERE_ALL = SRVF_BE_SPHERE
        AP_SPHERE_ALL = AP_BE_SPHERE
    else:
        PW_ROD_ALL = np.vstack((PW_ROD_ALL, PW_BE_ROD))
        Euclidean_ROD_ALL = np.vstack((Euclidean_ROD_ALL, Euclidean_BE_ROD))
        SRVF_ROD_ALL = np.vstack((SRVF_ROD_ALL, SRVF_BE_ROD))
        AP_ROD_ALL = np.vstack((AP_ROD_ALL, AP_BE_ROD))
        PW_SPHERE_ALL = np.vstack((PW_SPHERE_ALL, PW_BE_SPHERE))
        Euclidean_SPHERE_ALL = np.vstack((Euclidean_SPHERE_ALL, Euclidean_BE_SPHERE))
        SRVF_SPHERE_ALL = np.vstack((SRVF_SPHERE_ALL, SRVF_BE_SPHERE))
        AP_SPHERE_ALL = np.vstack((AP_SPHERE_ALL, AP_BE_SPHERE))
```

## Load Best Estimate Spectra

```python
def read_best_estimate(path):
    for itr in range(1,7):
        BE_spectra = pd.read_excel(path + str(itr) + '.xlsx')
        BE_array = np.array(BE_spectra)
        if itr == 1:
            BE_array_all = BE_array[:,1].reshape(1,-1)
        else:
            BE_array_all = np.vstack((BE_array_all, BE_array[:,1].reshape(1,-1)))
    return BE_array_all
```

```python
df = pd.read_excel('../Data_peakwv_SPHERE/Best_Estimate_1.xlsx')
Wavelength = df['Wavelength'].values
```

```python
PW_SPHERE_SPEC = read_best_estimate('../Data_peakwv_SPHERE/Best_Estimate_')
Euclidean_SPHERE_SPEC = read_best_estimate('../Data_Euclidean_SPHERE/Best_Estimate_')
SRVF_SPHERE_SPEC = read_best_estimate('../Data_SRVF_SPHERE/Best_Estimate_')
AP_SPHERE_SPEC = read_best_estimate('../Data_AP_SPHERE/Best_Estimate_')
PW_ROD_SPEC = read_best_estimate('../Data_peakwv_ROD/Best_Estimate_')
Euclidean_ROD_SPEC = read_best_estimate('../Data_Euclidean_ROD/Best_Estimate_')
SRVF_ROD_SPEC = read_best_estimate('../Data_SRVF_ROD/Best_Estimate_')
AP_ROD_SPEC = read_best_estimate('../Data_AP_ROD/Best_Estimate_')
```
